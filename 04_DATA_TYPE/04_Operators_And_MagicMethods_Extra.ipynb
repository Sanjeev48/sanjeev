```python
''' 
==============================================
            |Binary Operators|
==============================================
|Operator|                    |Method|                 
==============================================
    +        object.__add__(self, other)
    -        object.__sub__(self, other)
    *        object.__mul__(self, other)
    /        object.__div__(self, other)        #__future__.division is not in effect 
    /        object.__truediv__(self, other)    #__future__.division is in effect
    //       object.__floordiv__(self, other)
    %        object.__mod__(self, other)
    **       object.__pow__(self, other[, modulo])
    <<       object.__lshift__(self, other)
    >>       object.__rshift__(self, other)
    &        object.__and__(self, other)
    ^        object.__xor__(self, other)
    |        object.__or__(self, other) 

'''

''' 
To make division operator / work as __truediv__ we have to import divion from __future__ class,
like Shown in the below line:

from __future__ import division
'''
```

```python
'''
==============================================
           |Extended Assignments|
==============================================
|Operator|                   |Method|                 
==============================================
    +=       object.__iadd__(self, other)
    -=       object.__isub__(self, other)
    *=       object.__imul__(self, other)
    /=       object.__idiv__(self, other)       #__future__.division is not in effect
    /=       object.__itruediv__(self, other)    #__future__.division is in effect
    //=      object.__ifloordiv__(self, other)
    %=       object.__imod__(self, other)
    **=      object.__ipow__(self, other[, modulo])
    <<=      object.__ilshift__(self, other)
    >>=      object.__irshift__(self, other)
    &=       object.__iand__(self, other)
    ^=       object.__ixor__(self, other)
    |=       object.__ior__(self, other)

'''
```
